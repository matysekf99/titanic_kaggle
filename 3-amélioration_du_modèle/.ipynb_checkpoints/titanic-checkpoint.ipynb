### 2-modifier et nettoyer les données¶

```python
data_train = df_train.copy()
data_train
```

```python
data_train['Titre_civilité'] = df_train.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
data_train['Titre_civilité'].value_counts()
```

```python
data_train
```

```python
data_train['Titre_civilité'] = data_train['Titre_civilité'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Autre')
data_train['Titre_civilité'] = data_train['Titre_civilité'].replace('Mlle', 'Miss')
data_train['Titre_civilité'] = data_train['Titre_civilité'].replace('Ms', 'Miss')
data_train['Titre_civilité'] = data_train['Titre_civilité'].replace('Mme', 'Mrs')


Titre_civilité_groupby = data_train.groupby('Titre_civilité')['Survived'].mean().sort_values(ascending = False)
Titre_civilité_groupby
```

```python
Titre_civilité_age = sns.FacetGrid(data_train, row='Titre_civilité')
Titre_civilité_age.map(plt.hist, 'Age', bins=20)
Titre_civilité_age.add_legend()
Titre_civilité_age.set_ylabels('Count')
```

```python
Titre_civilité_groupby_age_mean = data_train.groupby('Titre_civilité')['Age'].mean()
Titre_civilité_groupby_age_mean
```

```python
Titre_civilité_groupby_age_median = data_train.groupby('Titre_civilité')['Age'].median()
Titre_civilité_groupby_age_median
```

```python
data_train['Age'] = data_train.apply(lambda x: Titre_civilité_groupby_age_median[x['Titre_civilité']] if pd.isnull(x['Age']) else x['Age'], axis=1)
data_train.head(30)
```

```python
data_train.info()
```

```python
data_train['Embarked'].value_counts()
```

```python
data_train.loc[data_train['Embarked'].isnull(),'Embarked']= 'S'
data_train.info()
```

```python
sns.catplot(x = 'Survived', col = 'Embarked', kind = 'count', data = data_train, palette=['blue', 'orange'],legend=False)
```

```python
data_train.loc[data_train['Cabin'].isnull(),'Cabin']= 'Non'
data_train.loc[data_train['Cabin']!='Non','Cabin']= 'Oui'
data_train.head()
```

```python
sns.catplot(x = 'Survived', col = 'Cabin', kind = 'count', data = data_train, palette=['blue', 'orange'],legend=False)
```

```python
data_train.info()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(df_train.corr(), annot = True, cmap='YlGnBu')
```

```python
data_train['A_une_cabine'] = 0
data_train.loc[(data_train['Cabin'] == 'Oui'), 'A_une_cabine'] = 1
data_train
```

```python
data_train
```

```python
data_train.drop(['Name','Cabin','Ticket'], inplace=True, axis=1)
data_train
```

```python
data_train['Sex'] = data_train['Sex'].replace({'male': 0, 'female': 1})
data_train['Embarked'] = data_train['Embarked'].replace({'S': 0, 'C': 1,'Q':2})
data_train['Titre_civilité'] = data_train['Titre_civilité'].replace({'Autre': 0, 'Master': 1, 'Miss': 2, 'Mr': 3, 'Mrs': 4})
data_train
```

```python
data_train['Accompagne'] = 0
data_train.loc[(data_train['Parch'] > 0) | (data_train['SibSp'] > 0), 'Accompagne'] = 1
data_train
```

```python
sns.catplot(x = 'Survived', col = 'Accompagne', kind = 'count', data = data_train, palette=['blue', 'orange'],legend=False)
```

```python
accompagne_groupby = data_train.groupby('Accompagne')['Survived'].mean()
accompagne_groupby
```

```python
data_train['Tranche_Age'] = pd.cut(data_train['Age'], 5)
data_train
```

```python
sns.catplot(x = 'Survived', col = 'Tranche_Age', kind = 'count', data = data_train, palette=['blue', 'orange'],legend=False)
```

```python
data_train.loc[ data_train['Age'] <= 16, 'Age'] = 0
data_train.loc[(data_train['Age'] > 16) & (data_train['Age'] <= 32), 'Age'] = 1
data_train.loc[(data_train['Age'] > 32) & (data_train['Age'] <= 48), 'Age'] = 2
data_train.loc[(data_train['Age'] > 48) & (data_train['Age'] <= 64), 'Age'] = 3
data_train.loc[ data_train['Age'] > 64, 'Age'] = 4
data_train['Age'] = data_train['Age'].astype(int)
data_train
```

```python
data_train['Tranche_Tarif']=pd.qcut(data_train['Fare'],4)
sns.catplot(x = 'Survived', col = 'Tranche_Tarif', kind = 'count', data = data_train, palette=['blue', 'orange'],legend=False)
```

```python
data_train.loc[ data_train['Fare'] <= 7.91, 'Fare'] = 0
data_train.loc[(data_train['Fare'] > 7.91) & (data_train['Fare'] <= 14.454), 'Fare'] = 1
data_train.loc[(data_train['Fare'] > 14.454) & (data_train['Fare'] <= 31), 'Fare']   = 2
data_train.loc[ data_train['Fare'] > 31, 'Fare'] = 3
data_train['Fare'] = data_train['Fare'].astype(int)
data_train
```

```python
data_train.drop(['SibSp','Parch','Tranche_Age','Tranche_Tarif'], inplace=True, axis=1)
data_train
```

### 3-comparer les modèles de prédiction

### Machine learning

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
```

```python
X_train = data_train.drop(['Survived','PassengerId'], axis=1)
Y_train = data_train['Survived']
X_test  = data_test.drop('PassengerId', axis=1)

display(X_train)
display(Y_train)
display(X_test)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred_logreg = logreg.predict(X_test)
logreg_score = logreg.score(X_train, Y_train)
logreg_score
```

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
Y_pred_random_forest = random_forest.predict(X_test)
random_forest_score = random_forest.score(X_train, Y_train)
random_forest_score
```

```python
svc = SVC()
svc.fit(X_train, Y_train)
Y_pred_svc = svc.predict(X_test)
svc_score = svc.score(X_train, Y_train)
svc_score
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)
Y_pred_knn = knn.predict(X_test)
knn_score = knn.score(X_train, Y_train)
knn_score
```

```python
mlp_model = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000)
mlp_model.fit(X_train, Y_train)
Y_pred_mlp = mlp_model.predict(X_test)
mlp_score = mlp_model.score(X_train, Y_train)
mlp_score
```

```python
tableau_de_comparaison = {
    'Modèle': ['Régression Logistique', 'Forêts Aléatoires', 'SVM', 'k-NN', 'Réseaux de Neurones'],
    'Score': [logreg_score, random_forest_score, svc_score, knn_score, mlp_score]
}
tableau_de_comparaison = pd.DataFrame(tableau_de_comparaison)
tableau_de_comparaison.sort_values(by='Score', ascending=False)
```

```python
resultat = pd.DataFrame({'PassengerId': df_test['PassengerId'],'Survived': Y_pred_mlp})
resultat.to_csv('solutions/resultat_neurones.csv', index=False)
```

```python
resultat = pd.DataFrame({'PassengerId': df_test['PassengerId'],'Survived': Y_pred_knn})
resultat.to_csv('solutions/resultat_knn.csv', index=False)
```

```python
resultat = pd.DataFrame({'PassengerId': df_test['PassengerId'],'Survived': Y_pred_random_forest})
resultat.to_csv('solutions/resultat_random_forest.csv', index=False)
```

```python
resultat = pd.DataFrame({'PassengerId': df_test['PassengerId'],'Survived': Y_pred_svc})
resultat.to_csv('solutions/resultat_svc.csv', index=False)
```
