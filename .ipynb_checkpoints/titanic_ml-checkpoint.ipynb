## 4-comparer les modèles de prédiction

```python
X_train = df_train.drop('Survived', axis=1)
Y_train = df_train['Survived']
X_test  = df_test.drop('PassengerId', axis=1)

display(X_train)
display(Y_train)
display(X_test)
```

#### Régression logistique :

Type de modèle : Classification   
Description : La régression logistique est utilisée pour résoudre des problèmes de classification binaire,    
comme la prédiction de la survie (1) ou la non-survie (0) des passagers. Elle est relativement simple et efficace

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred_logreg = logreg.predict(X_test)
logreg_score = logreg.score(X_train, Y_train)
logreg_score
```

#### Forêts aléatoires (Random Forest) :

Type de modèle : Ensembliste (ensemble method)   
Description : Les forêts aléatoires sont constituées de multiples arbres de décision.   
Chaque arbre vote pour la classe de prédiction, et la classe avec le plus de votes est sélectionnée.   
Elles sont robustes et performantes, adaptées à des ensembles de données complexes.   

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
Y_pred_random_forest = random_forest.predict(X_test)
random_forest_score = random_forest.score(X_train, Y_train)
random_forest_score
```

#### Support Vector Machines (SVM) :

Type de modèle : Classification   
Description : Les SVM cherchent à trouver un hyperplan optimal pour séparer les classes.   
Ils sont efficaces dans des espaces de grande dimension et peuvent gérer à la fois des problèmes de classification linéaire et non linéaire.   

```python
svc = SVC()
svc.fit(X_train, Y_train)
Y_pred_svc = svc.predict(X_test)
svc_score = svc.score(X_train, Y_train)
svc_score
```

#### k-Nearest Neighbors (k-NN) :

Type de modèle : Classification   
Description : Le k-NN attribue une classe en fonction de la majorité des classes des k voisins les plus proches dans l'espace des caractéristiques.   
C'est simple à comprendre, mais peut être sensible à l'échelle des caractéristiques.   

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)
Y_pred_knn = knn.predict(X_test)
knn_score = knn.score(X_train, Y_train)
knn_score
```

#### Réseaux de neurones (Neural Networks) :

Type de modèle : Classification   
Description : Les réseaux de neurones sont des modèles puissants qui peuvent apprendre des représentations complexes.   
Ils peuvent être utilisés pour des tâches de classification, mais peuvent nécessiter un réglage minutieux des hyperparamètres.   

```python
mlp_model = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000)
mlp_model.fit(X_train, Y_train)
Y_pred_mlp = mlp_model.predict(X_test)
mlp_score = mlp_model.score(X_train, Y_train)
mlp_score
```

```python
tableau_de_comparaison = {
    'Modèle': ['Régression Logistique', 'Forêts Aléatoires', 'SVM', 'k-NN', 'Réseaux de Neurones'],
    'Score': [logreg_score, random_forest_score, svc_score, knn_score, mlp_score]
}
tableau_de_comparaison = pd.DataFrame(tableau_de_comparaison)
tableau_de_comparaison.sort_values(by='Score', ascending=False)
```
